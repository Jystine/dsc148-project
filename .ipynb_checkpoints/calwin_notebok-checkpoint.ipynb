```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.svm import SVC

from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# what to do:
# knn svm logistic random_forest 
# us / not us 
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
adult = fetch_ucirepo(id=2) 
  
# data (as pandas dataframes) 
X = adult.data.features 
y = adult.data.targets 
```

```python
# consolidating income col values
y[y['income'] == '<=50K.'] = '<=50K'
y[y['income'] == '>50K.'] = '>50K'
y['income'] = y['income'].apply(lambda x: 1 if x == '<=50K' else 0)
```

#### Missing Values

```python
# finds all rows that has a "?" or missing values, then filters dataframe to get rows that do not have those values
cols = X.columns
has = ~(X['age'] == X['age'])
for col in cols:
    has = ((has) | (X[col] =='?') | (X[col].isna()))

X = X[~has]
y = y[~has]
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Logistic Regression Model

```python
preproc = ColumnTransformer(
    transformers = [
        ("categorical", OneHotEncoder(handle_unknown='ignore', sparse_output = False), ["race", "sex", "occupation"]),
        ("numerical", "passthrough", ["age", "hours-per-week", "education-num"])
    ],
    remainder = "drop"
)
```

```python
plr = Pipeline([
    ("preproc", preproc),
    ("model", LogisticRegression())
])

plr.fit(X_train, y_train)
```

```python
predicted = plr.predict(X_test)
```

```python
# accuracy
(pd.Series(predicted) == y_test.reset_index()['income']).mean()
```

```python
# recall
recall_score(y_test, predicted)
```

```python
# f1 score
f1_score(y_test, predicted)
```

```python
# AUC-ROC
roc_auc_score(y_test, predicted)
```

# Random Forrest

# LightGBM

```python
from lightgbm import LGBMClassifier

```

```python
categorical = ["race", "sex", "occupation", "workclass", "education", "marital-status", "relationship", "native-country"]
preproc = ColumnTransformer(
    transformers = [
        ("categorical", OneHotEncoder(handle_unknown='ignore', sparse_output = True), categorical),
        ("numerical", "passthrough", ["age", "hours-per-week", "fnlwgt", "capital-gain", "capital-loss", "education-num"])
    ],
    remainder = "drop"
)
```

```python
plr = Pipeline([
    ("preproc", preproc),
    ("model", LGBMClassifier(max_depth = 10, learning_rate = .01, n_estimators = 1000, n_jobs = -1))
])
plr.fit(X_train, y_train)
```

```python
predicted = plr.predict(X_test)
```

```python
# accuracy
(pd.Series(predicted) == y_test.reset_index()['income']).mean()
```

```python
# recall
recall_score(y_test, predicted)
```

```python
# f1 score
f1_score(y_test, predicted)
```

```python
# AUC-ROC
roc_auc_score(y_test, predicted)
```
